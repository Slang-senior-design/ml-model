```python
import pandas as pd
columns = ["pixel"+str(i) for i in range(10000)]
df = pd.read_csv("mapping.csv", names=["class"]+columns, dtype={'class':str, 'img':str})
df.head()
```

```python
df["class"].unique()
```

```python
alone = df[df["class"] =="alone"]
twenty = df[df["class"] == "twenty"]
lonely = df[df["class"] == "lonely"]
```

```python
alone.describe()
```

```python
twenty.describe()
```

```python
lonely.describe()
```

```python
alone_imgs = alone.loc[:, df.columns != 'class']
```

```python
imgs = [alone_imgs.iloc[i].values.reshape(100,100) for i in range(alone_imgs.shape[0])]
```

```python
import cv2
import numpy as np
avg = np.mean(imgs, axis=1)
avg
```

```python
cv2.imshow("alone", avg)
cv2.waitKey()
```
